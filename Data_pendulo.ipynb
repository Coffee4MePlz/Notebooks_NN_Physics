```python
import pandas as pd

# Substitua 'caminho_para_seu_arquivo.csv' pelo caminho at√© o seu arquivo
df = pd.read_csv('Raw Data.csv')

# Exibir as primeiras linhas do DataFrame
print(df.head())
```

```python
# Converter colunas para NumPy arrays
time = df['Time (s)'].to_numpy()
x_data = df['Rotation x (rad/s)'].to_numpy()
y_data = df['Rotation y (rad/s)'].to_numpy()
z_data = df['Rotation z (rad/s)'].to_numpy()
# Criar gr√°ficos para visualizar os dados
plt.figure(figsize=(10, 3))

# Gr√°fico da coluna A
plt.subplot(1, 3, 1)  # 1 linha, 3 colunas, posi√ß√£o 1
plt.plot(time,x_data, label='x_data')
plt.title('Coluna A')
plt.legend()

# Gr√°fico da coluna B
plt.subplot(1, 3, 2)  # 1 linha, 3 colunas, posi√ß√£o 2
plt.plot(time,y_data, label='y_data,', color='green')
plt.title('Coluna B')
plt.legend()

# Gr√°fico da coluna C
plt.subplot(1, 3, 3)  # 1 linha, 3 colunas, posi√ß√£o 3
plt.plot(time,z_data, label='z C', color='red')
plt.title('Coluna C')
plt.legend()

# Exibir o gr√°fico
plt.tight_layout()
plt.show()
```

# Coordenadas polares

```python
def eq_movimento(theta_0,omega_0,t_):
    return  theta_0*np.cos(omega_0*t_)

t = np.linspace(0,0.5*np.pi)
theta_t = eq_movimento(theta_0=np.pi/6,omega_0=np.sqrt(9.81/0.5),t_=t)

plt.plot(t,theta_t,".--",label="$ theta$")
# Adicionando r√≥tulos e t√≠tulo
plt.xlabel("Tempo (s)")
plt.ylabel("√Çngulo (rad) ")
plt.title("Movimento do P√™ndulo Simples em Coordenadas Polares")
plt.legend()
plt.grid(True)
plt.show()
plt.show()
```

# Coordenadas Carteziana

```python
def eq_movimento_x(l,theta):
    return  l*np.sin(theta) , -l*np.cos(theta) 

#t = np.linspace(0,1*np.pi)
x,y = eq_movimento_x(l=0.5,theta=theta_t)

plt.plot(t,x,"-",label="x")
plt.plot(t,y,"-",label="y")

# Adicionando r√≥tulos e t√≠tulo
plt.xlabel("√Çngulo (rad)")
plt.ylabel("Posi√ß√£o")
plt.title("Movimento do P√™ndulo Simples em Coordenadas Cartesianas")
plt.legend()
plt.grid(True)
plt.show()
```
