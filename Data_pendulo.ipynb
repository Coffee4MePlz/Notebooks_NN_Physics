# Coordenadas polares

```python
def eq_movimento(theta_0,omega_0,t_):
    return  theta_0*np.cos(omega_0*t_)

t = np.linspace(0,0.5*np.pi)
theta_t = eq_movimento(theta_0=np.pi/6,omega_0=np.sqrt(9.81/0.5),t_=t)

plt.plot(t,theta_t,".--",label="$ theta$")
# Adicionando rótulos e título
plt.xlabel("Tempo (s)")
plt.ylabel("Ângulo (rad) ")
plt.title("Movimento do Pêndulo Simples em Coordenadas Polares")
plt.legend()
plt.grid(True)
plt.show()
plt.show()
```

# Coordenadas Carteziana

```python
def eq_movimento_x(l,theta):
    return  l*np.sin(theta) , -l*np.cos(theta) 

#t = np.linspace(0,1*np.pi)
x,y = eq_movimento_x(l=0.5,theta=theta_t)

plt.plot(theta_t,x,"-",label="x")
plt.plot(theta_t,y,"-",label="y")

# Adicionando rótulos e título
plt.xlabel("Ângulo (rad)")
plt.ylabel("Posição")
plt.title("Movimento do Pêndulo Simples em Coordenadas Cartesianas")
plt.legend()
plt.grid(True)
plt.show()
```
